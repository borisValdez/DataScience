```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_diabetes = df_dataset[df_dataset['Disease'] == 'Diabetes']
df_sano = df_dataset[df_dataset['Disease'] == 'Healthy']
```

```python
indice = df_diabetes.index
df_diabetes
indice2=df_sano.index
```

```python
variables_y = ['Glucose', 'Cholesterol']

# Crear el gráfico lineal
plt.figure(figsize=(20, 6))
for variable in variables_y:
    plt.plot(indice, df_diabetes[variable], label=variable)

# Añadir leyenda y etiquetas
plt.legend()
plt.xlabel('Índice del DataFrame')
plt.ylabel('Valor')
plt.title('Variables para casos de Diabetes')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
variables_y = ['Glucose', 'Cholesterol']

# Crear el gráfico lineal
plt.figure(figsize=(20, 6))
for variable in variables_y:
    plt.plot(indice2, df_sano[variable], label=variable)

# Añadir leyenda y etiquetas
plt.legend()
plt.xlabel('Índice del DataFrame')
plt.ylabel('Valor')
plt.title('Variables para casos de Sanos')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

Se puede observar que los pacientes con diabetes, llegan a valores mas altos tanto de glucosa como de colesterol

```python
plt.figure(figsize=(20, 6))
colors = ['red', 'blue']
plt.scatter(indice, df_diabetes['Cholesterol'], color='blue', marker='o', label= 'Diabetes')
plt.scatter(indice2, df_sano['Cholesterol'], color='red', marker='o', label="Sanos")
# Añadir título y etiquetas de los ejes
plt.title('Comparación colesterol en casos sanos y casos con diabetes')
plt.xlabel('Paciente')
plt.ylabel('Colesterol')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
colors = ['red', 'blue']
plt.scatter(indice, df_diabetes['Glucose'], color='blue', marker='o')
plt.scatter(indice2, df_sano['Glucose'], color='red', marker='o')
# Añadir título y etiquetas de los ejes
plt.title('Comparación Glucosa en casos sanos y casos con diabetes')
plt.xlabel('Paciente')
plt.ylabel('Glucosa')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
