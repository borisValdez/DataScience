```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_diabetes = df_dataset[df_dataset['Disease'] == 'Diabetes']
df_sano = df_dataset[df_dataset['Disease'] == 'Healthy']
```

```python
indice = df_diabetes.index
df_diabetes
indice2=df_sano.index
```

```python
variables_y = ['Glucose', 'Cholesterol']

# Crear el gráfico lineal
plt.figure(figsize=(20, 6))
for variable in variables_y:
    plt.plot(indice, df_diabetes[variable], label=variable)

# Añadir leyenda y etiquetas
plt.legend()
plt.xlabel('Índice del DataFrame')
plt.ylabel('Valor')
plt.title('1.1 - Variables para casos de Diabetes')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

Para las variables de Glucosa se utilizara el color azul y/o celeste, en cambio, para las variables de Colesterol se utilizara naraja y/o rojo.

```python
variables_y = ['Glucose', 'Cholesterol']

# Crear el gráfico lineal
plt.figure(figsize=(20, 6))
for variable in variables_y:
    plt.plot(indice2, df_sano[variable], label=variable)

# Añadir leyenda y etiquetas
plt.legend()
plt.xlabel('Índice del DataFrame')
plt.ylabel('Valor')
plt.title('1.2 - Variables para casos de Sanos')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

Conclusión graficos 1.1 y 1.2: Se puede observar que los pacientes con diabetes, llegan a valores mas altos tanto de glucosa, como de colesterol.
En el caso de la Glucosa, los pacientes con diabetes se pueden observar picos donde llegan casi a "1", en cambio en pacientes sanos este valor no suele superar el "0.7".
Por otro lado, se puede apreciar que los pacientes con diabetes llegan a valores de colesterol que tocan casi el "0.9", en cambio los pacientes sanos no suelen superar el "0.7".
Estos graficos nos dan un primer indicio de que hay determinada relación directa entre la glucosa y el colesterol que podrian servir de indicio para diagnosticar la diabetes.

```python
plt.figure(figsize=(20, 6))
colors = ['red', 'blue']
plt.scatter(indice, df_diabetes['Cholesterol'], color='orange', marker='o')
plt.scatter(indice2, df_sano['Cholesterol'], color='red', marker='o')
# Añadir título y etiquetas de los ejes
plt.title('Comparación colesterol en casos sanos y casos con diabetes')
plt.xlabel('Paciente')
plt.ylabel('Colesterol')
plt.legend(["Con diabetes","Sanos"])

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

En este gráfico se termina de evidenciar que los pacientes con diabetes (naranja) llegan a un nivel mayor de colesterol que los pacientes sanos,marcando un límite cerca de "0.7".

```python
plt.figure(figsize=(15, 6))
colors = ['red', 'blue']
plt.scatter(indice, df_diabetes['Glucose'], color='blue', marker='o')
plt.scatter(indice2, df_sano['Glucose'], color='skyblue', marker='o')
# Añadir título y etiquetas de los ejes
plt.title('Comparación Glucosa en casos sanos y casos con diabetes')
plt.xlabel('Paciente')
plt.ylabel('Glucosa')
plt.legend(["Con diabetes", "Sanos"])

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

En este gráfico se puede evidenciar que existe un límite en los valores de colesterol para los pacientes sanos (Celeste), siendo este de "0.8", solo los pacientes con diabetes superan ese valor.
