Se procederá a realizar un modelo de machine learning de clustering, para poder ver como se pueden categorizar los tipos de pacientes

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=3000, centers=4, cluster_std=0.70, random_state=0)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Metodo del codo')
plt.xlabel('Numero de clusters')
plt.ylabel('Inercia')
plt.show()
```

Según la prueba del codo, puedo ver que lo mejor es generar 3 o 4 clusters.

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
X = df_diabetes[['Cholesterol', 'Glucose']].values
model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
model.fit(X)
labels = model.labels_
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, marker='o', color='red')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, marker='o', color='blue')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=50, marker='o', color='green')
plt.scatter(X[labels==3, 0], X[labels==3, 1], s=50, marker='o', color='purple')
plt.show()

X = df_sano[['Cholesterol', 'Glucose']].values
model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
model.fit(X)
labels = model.labels_
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, marker='o', color='red')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, marker='o', color='blue')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=50, marker='o', color='green')
plt.scatter(X[labels==3, 0], X[labels==3, 1], s=50, marker='o', color='purple')
plt.show()
```

```python
X = df_dataset[['Cholesterol', 'Glucose']].values
model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
# Crear el modelo de clustering
model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')

# Entrenar el modelo
model.fit(X)

# Obtener las etiquetas de los clusters
labels = model.labels_

# Graficar los clusters
plt.scatter(X[labels==0, 0], X[labels==0, 1], s=50, marker='o', color='red', label='Cluster 0 (Disease)')
plt.scatter(X[labels==1, 0], X[labels==1, 1], s=50, marker='o', color='blue', label='Cluster 1 (Disease)')
plt.scatter(X[labels==2, 0], X[labels==2, 1], s=50, marker='o', color='green', label='Cluster 2 (Disease)')
plt.scatter(X[labels==3, 0], X[labels==3, 1], s=50, marker='o', color='purple', label='Cluster 3 (Disease)')

# Obtener las coordenadas para los pacientes con diabetes
diabetes_data = df_dataset[df_dataset['Disease'] == 1]
diabetes_chol = diabetes_data['Cholesterol'].values
diabetes_gluc = diabetes_data['Glucose'].values

# Graficar los puntos correspondientes a pacientes con diabetes
plt.scatter(diabetes_chol, diabetes_gluc, s=100, marker='x', color='black', label='Diabetes')

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.xlabel('Cholesterol')
plt.ylabel('Glucose')
plt.title('Distribución de pacientes con Clustering')
plt.show()
```
